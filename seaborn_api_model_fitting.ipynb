```python
f, ax = plt.subplots(figsize=(12, 12))
sns.regplot(x='BldgArea',
            y='SalePrice',
            data=sales[(10000 < sales['BldgArea']) & (sales['BldgArea'] < 0.1*10**7) & (sales['SalePrice'] < 1*10**9)],
            scatter_kws={'edgecolor': "white",
                         'linewidth': 1},

          )
ax.set_xlim([0, 1000000])
ax.set_ylim([0, 7*10**8])
```

This dataset doesn't have a column that needs it really but also check out `x_jitter=.05` and `x_estimator=np.mean` parameters.

That was linear, how about polynomial!

```python
f, ax = plt.subplots(figsize=(12, 12))
sns.regplot(x='BldgArea',
            y='SalePrice',
            data=sales[(10000 < sales['BldgArea']) & (sales['BldgArea'] < 0.1*10**7) & (sales['SalePrice'] < 1*10**9)],
            scatter_kws={'edgecolor': "white",
                         'linewidth': 1},
            order=2
          )
ax.set_xlim([0, 1000000])
ax.set_ylim([0, 7*10**8])
```

Interesting; false, obviously, but observationally interesting. The paucity of data on the outer half causes this.

Note that linear regression also has a `robust=True` option to downweigh extreme outliers.

Enable logistic regression with `logistic=True`, here we plot a logistic regression for a dummy variable for buildings worth over 10 million.

```python
log_dummy = sales[(sales['SalePrice'] > 1000) &
                  (sales['SalePrice'] < 1*10**9) &
                  (10000 < sales['BldgArea']) &
                  (sales['BldgArea'] < 0.1*10**6)]
log_dummy['Mil'] = log_dummy['SalePrice'] > 10000000
```

```python
sns.regplot(x='BldgArea',
            y='Mil',
            data=log_dummy,
            scatter_kws={'edgecolor': "white",
                         'linewidth': 1},
            logistic=True
          )
```

```python
sns.set(font_scale=1)
sns.lmplot(x="BldgArea", y="SalePrice", hue="OwnerType", data=sales[sales['SalePrice'] < 1*10**9])
```

Looking at the distribution it's really just "Private" or "Not Private" that we're considering here.

```python
owner_dummy = sales[sales['SalePrice'] < 1*10**9].copy()
owner_dummy['P/NP'] = (owner_dummy['OwnerType'] == 'P').replace(True, "P").replace(False, "NP")
```

```python
sns.set(font_scale=1)
sns.lmplot(x="BldgArea", y="SalePrice", hue='P/NP', data=owner_dummy, markers=["o", "x"], palette="Set1")
```

This is the single-plot approach, you can also use the small multiples approach using `col`.

```python
sns.lmplot(x="BldgArea", y="SalePrice", col='P/NP', data=owner_dummy)
```
